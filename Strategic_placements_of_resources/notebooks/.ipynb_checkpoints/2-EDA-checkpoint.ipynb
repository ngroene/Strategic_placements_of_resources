### Dynamic Time Warping (DTW)  
- one of the algorithms for measuring the similarity between two temporal time series sequences, which may vary in speed.

```python
weekly_sales_per_dept_df = master_df.groupby(['Dept', 'Date'])['Weekly_Sales'].sum().reset_index(name = 'Weekly_Sales_per_dept')
weekly_sales_per_dept_df.head()
```

```python
weekly_sales_per_dept_series = []
for dept in weekly_sales_per_dept_df['Dept'].unique().tolist():
    df_sub = weekly_sales_per_dept_df[weekly_sales_per_dept_df['Dept']==dept]
    sales_series_sub = df_sub['Weekly_Sales_per_dept'].to_numpy()
    weekly_sales_per_dept_series.append(sales_series_sub)
        
#     break
# sales_series_sub   
# weekly_sales_per_dept_series
    
```
